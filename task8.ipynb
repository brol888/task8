#3 მოცემულია JSON მასივი:
[
{"id": 1, "price": 50},
{"id": 2, "price": 200},
{"id": 3, "price": 150}
]
ამოიღე მხოლოდ ის პროდუქტები, რომელთა ფასი 100-ზე მეტია.

```python
import json

json_data = '''[
{"id": 1, "price": 50},
{"id": 2, "price": 200},
{"id": 3, "price": 150}
] '''

python_obj = json.loads(json_data)

python_obj = [i for i in python_obj if i['price'] > 100]

print(python_obj)
```
